```python
def change_rank(rank):
    if rank == "동일":
        rank = 0
    elif rank[-2:] == "하락":
        rank = int(rank[:-2]) * -1
    elif rank[-2:] == "상승":
        rank = int(rank[:-2]) * 1
    return rank

daily_boxoffice_6['순위 변동'] = daily_boxoffice_6['순위 변동'].map(lambda x: change_rank(x))


daily_boxoffice_6.head()
```
